# Logistic Regression

## Why

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.simplefilter("ignore")
```

描述一个事件的可能性，除了用概率以外，还可以用几率，胜率，赔率等等。

```python
p = np.linspace(0,1,100)
odds = p/(1-p)
logodds = np.log(odds)
```

```python
plt.plot(p, odds, label="odds")
plt.xlabel("p")
plt.ylabel("odds")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.semilogy(p, odds, label="odds")
plt.xlabel("p")
plt.ylabel("odds")
plt.legend()
plt.title("semilogy odds")
plt.grid()
plt.show()
```

```python
plt.plot(p, np.log10(odds),label="log10odds")
plt.xlabel("p")
plt.ylabel("log10odds")
plt.legend()
plt.title("log10odds")
plt.grid()
plt.show()
```

底数不是特别重要，但是合适的底方便对比。

```python
plt.plot(p, np.log10(odds),label="log10odds")
plt.vlines(0.5, -2, 2, color="g", linestyle="dashed")
plt.hlines(0, 0.4, 0.6, color="g", linestyle="dashed")
plt.text(0.5, 0, "(0.5, 0)")
plt.xlabel("p")
plt.ylabel("log10odds")
plt.legend()
plt.title("log10odds")
plt.grid()
plt.show()
```
